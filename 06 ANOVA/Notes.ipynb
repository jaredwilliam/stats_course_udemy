**Analysis of Variance**

In the previous section we used Z- and t-Distributions to answer qustions like *"What is the probability that two samples come from the same population?"*

In this section we introduce a new distribution - the F-Distribution



It is used to answer the question *"What is the probability that two samples come from populations that have the same variance?"*. It can also answer: *"What is the probability that three or more samples come from the same population?"*

# ANOVA - Analysis of Variance

In the previous section we tested two samples to see if they likely came from the same population. What if we have three or more samples? Could we do the same thing?



Our null hypothesis would look like:

<center>$H_0:\mu_A = \mu_B = \mu_C$</center>

We could test each pair but our overall confidence drops.



This is where ANOVA comes in. 

We compute an F value and compare it to a critical value determined by our degrees of freedom (the number of groups, and the number of items in each group)

Let's work with some data



First calculate the individual means:

A = 44, B = 50, C = 53

Then calculate the overall mean = 49

ANOVA consideres two types of variance:
- Between groups: how far group means stray from the total mean
- Within groups: how far inidividual values stray from thier respective group mean.

The F-value we are trying to calculate is simply the ratio between these two variances!

<center>$F = \frac{Variance\;Between\;Groups}{Variance\;Within\;Groups}$</center>





We find that F = 1.718

# F Distribution



Look up our critical value from an F-table. 



Since our calculated value is less than the critical value, we fail to reject the null hypothesis.
