# Probability

## Permutations

A **permutation** of a set of objects is an arragement of the objects in a certain order. The possible of letter **a**, **b**, and **c** are:

- abc
- acb
- bac
- bca
- cab
- cba

For simple examples like **abc**, we calculate the number of possible permutations as $n!$ (n factorial). 

$n! = 3! = 3 * 2 * 1 = 6$

You can also take a subset of items in a permutation. The number of permutation of a set of $n$ objects taken $r$ at a time is given by the following formula:

<center>$_nP_r = \frac{n!}{(n-r)!}$</center>

For example, if you were trying to figure out how many permutations of three letters you can have out of the entire 26-letter alphabet, $n=26$ and $r=3$.

Image you have a website that requires a four character password. Characters can either be lowercase letters of the digits 0-9. You may not repeat a letter or number. How many different possible passwords can there be?

```python
import math
r = 4
n = 26 + 10 
possible = math.factorial(n) / math.factorial(n-r)
print(possible)
```

What if we did allow repitition?

The number of arrangement of $n$ objects taken $r$ at a time, *with repitition*, is given by:

<center>$n^r$</center>

How many 4 digit license plates can you make using the numbers 0 to 9 while allowing repitition?

```python
n = 10
r = 4
possible = n**r
print(possible)
```

## Combinations

Combinations are *unordered* arrangements of objects. A group of people selected for a team are the same group, no matter the order. 

The number of combinations of a set of $n$ objects take $r$ at a time is given by:

<center> $_nC_r = \frac{n!}{r!(n-r)!}$ </center>

How many 3-letter combinations cna be made from the letters ABCDE?

Let's also compare this with the number of permutations you have. The number of perumuations is 60. But only 10 different combinations. 

For a study, 4 people are chosen at random from a group of 10 people. How many ways can this be done?

<center>$_nC_r = \frac{n!}{r!(n-r)!} = \frac{10!}{4!(10-4)!} = 210$</center>

What if we were allowed to repeat choices? In that case, that is combination with repetition.



**Combinations with Repitition**

The number of combinations taken r at a time from a set n and allowing for repetition:

<center>$_nC_r = \frac{(n+r-1)!}{r!(n-1)!}$</center>

For a pizza, 4 ingredients are chosen at random from a possible of 10 ingredients. How many different pizza topping combinations are there, allowing repitition?

<center>$_nC_r = \frac{(n+r-1)!}{r!(n-1)!} = \frac{(10+4-1)!}{4!(10-1)!} = 715$</center>

# Intersections, Unions, and Complements

In probability, an **intersection** describes the sample space where two events *both* occur. Consider a box of patterned, colored balls. 

```python
Image(filename='red and striped balls.png')
```

What are the odds of a red, striped ball? If we assign **A** as the even of red balls and **B** as the event of striped balls, the intersection of **A *and* B** is given as:

<center>$A \cap B$</center>

Also, the order doesn't matter, 

<center>$A \cap B = B \cap A $</center>

The probability and A *and* B is given as:

<center>$P(A \cap B)$</center>

In the case of our red and striped balls, the probability of grabbing a red striped ball is:

<center>$P(A \cap B) = \frac{3}{15} = 0.2$</center>

The **union** of two events considers if **A *or* B** occurs, and is given by:

<center>$A \cup B$</center>

Order doesn't matter:

<center>$A \cup B = B \cup A$</center>

The probability of A *or* B is given as:

<center>$P(A \cup B) = P(A) + P(B) - P(A \cap B)$</center>

So for our example:

<center>$P(A \cup B) = \frac{9}{15} + \frac{9}{15} - \frac{3}{15} = \frac{15}{15} = 1.0$</center>

The **complement** of an event considers everything outside of the event, given by:

<center>$\overline{A}$</center>
    
The probability of *not* A is:
    
<center>$P( \overline{A}) = 1 - P(A) = \frac{15}{15} - \frac{9}{15} = \frac{6}{15} = 0.4 $</center>

# Independent and Dependent Events

An **independent** series of events occur when the outcome of one event has no effect on the outcome of another. An example is flipping a fair coin twice. The change of getting heads on the second toss is independent of the result of the first toss. 

The probability of seeing two heads with two flips of a fair coin is:

<center>$P(H_1H_2)=P(H_1) \times P(H_2) = \frac{1}{2} \times \frac{1}{2} = \frac{1}{4}$</center>

A **dependent** event occurs when the outcome of a first event does affect the probability of a second event. A common example is to draw colored marbles from a bag *without replacement*. 

Imagine a bag with three red marbles and two blue marbles. You take two marbles out of the bag without replacement. What is the probability they are both red?

<center>$P(R_1) = \frac{3}{5}$</center>

<center>$P(R_2|R_1) = \frac{2}{4}$</center>

<center>$P(R_1 \cap R_2) = P(R_1) \times P(R_2)$</center>

<center>$P = \frac{3}{5} \times \frac{2}{4} = 0.3$</center>

# Conditional Probability

The idea that we want to know the probability of event A, *given* that event B has occured, is **conditional probability**. This is written as:

<center>$P(A|B)$</center>

Lets go back to the two marbles. 

<center>$P(R_1 \cap R_2) = P(R_1) \times P(R_2|R_1)$</center>

The conditional in this equation is the second term:

<center>$P(R_2|R_1)$</center>

Rearranging the formula gives:

<center>$P(A|B) = \frac{P(A \cap B)}{P(B)}$</center>

That is, the probability of **A given B** equals the probability of **A and B** divided by the probability of **B**.

A company finds that out of every 100 projects, 48 are completed on time, 62 are completed under budget, and 16 are completed both on time and under budget. Given that a project is completed on time, what is the probability that is it under budget?

<center>$=\frac{16}{48}=0.33$</center>

# Addition and Multiplication Rules

When two events cannot both happen, they are said to be **mutually exclusive**. 

<center>$P(A \cup B) = P(A) + P(B)$</center>

The **multiplication rule**:

<center>$P(A \cap B) = P(A) \times P(B|A)$</center>

Given a standard deck of 52 cards, what is the probability of drawing four aces?

<center>$P(A \cap B \cap C \cap D) = P(A) \times P(B|A) \times (C|AB) \times P(D|ABC)$</center>

<center>$= \frac{4}{52} \times \frac{3}{51} \times \frac{2}{50} \times \frac{1}{49} = \frac{1}{270,725}$</center>

# Bayes Theorem

We have already seen conditional probability:

<center>$P(A|B) = \frac{P(A \cap B)}{P(B)}$</center>

Provided that P(B)>0. 

And

<center>$P(B|A) = \frac{P(B \cap A)}{P(A)} = \frac{P(A \cap B)}{P(A)}$</center>

Provided that P(A)>0.

We can use substutution to combine these equations together, that way we can relate these two probabilities. This is Baye's Theorem. 

<center>$P(A|B) =  \frac{P(B|A)P(A)}{P(B)}$</center>

Provided that P(A), P(B) >0

Bayes theorem is used to determing the probability of a parameter, given a certain event.
